```python
import pandas as pd
import sqlite3

df = pd.read_csv("nba_clean.csv")
print("Data loaded!")
```

```python
conn = sqlite3.connect("nba.db")

df.to_sql("players", conn, if_exists="replace", index=False)
print("Database created! Table 'players' is ready.")
```

```python
query = """
SELECT player_name, pts, reb, ast
FROM players
LIMIT 10
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT player_name, ROUNd(AVG(pts), 2) AS avg_pts
FROM players
GROUP BY player_name
ORDER BY avg_pts DESC
LIMIT 10
"""
result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT player_name, pts, reb, ast, season
FROM players
WHERE pts >= 25
ORDER BY pts DESC
"""
result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT team_abbreviation, 
ROUND(AVG(pts), 2) AS avg_pts,
ROUND(AVG(reb), 2) AS avg_reb,
ROUND(AVG(ast), 2) AS avg_ast
FROM players
GROUP BY team_abbreviation
ORDER BY avg_pts DESC
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT player_name, 
COUNT(season) AS seasons_played
FROM players
GROUP BY player_name
ORDER BY seasons_played DESC
LIMIT 15
"""
result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT player_name, 
ROUND(AVG(pts), 2) AS avg_pts,
ROUND(AVG(reb), 2) AS avg_reb,
ROUND(AVG(ast), 2) AS avg_ast,
ROUND(AVG(pts + reb + ast), 2) AS total_contribution
FROM players
GROUP BY player_name
ORDER BY total_contribution DESC
LIMIT 10
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
conn.close()
print("Database connection closed.")
```
