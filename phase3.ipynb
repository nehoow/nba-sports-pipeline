```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("nba_clean.csv")

sns.set_theme(style="darkgrid")
print("Ready to Visualize!")
```

```python
top_scorers = df.groupby("player_name")["pts"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 5))
top_scorers.plot(kind="bar", color="steelblue")
plt.title("Top 10 NBA Scorers (career average)")
plt.xlabel("Player")
plt.ylabel("Average Points per Game")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
player_2022 = df[df["season"] == "2021-22"][["player_name", "age", "pts"]]
plt.scatter(player_2022["age"], player_2022["pts"], alpha=0.3, color="coral")
plt.title("Age vs Points Scored")
plt.xlabel("Age")
plt.ylabel("Points Scored")
plt.show()
```

```python
season_pts = df.groupby("season")["pts"].mean()
plt.figure(figsize=(12,6))
season_pts.plot(kind="line", marker="o", color="purple")
plt.title("Average NBA Points per Season")
plt.xlabel("Season")
plt.ylabel("Avg points per game")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df["pts"].plot(kind="hist", bins=30, color="teal", edgecolor="white")
plt.title("Distribution of Points")
plt.xlabel("Points Scored")
plt.ylabel("Number of Players")
plt.show()
```

```python
stats = df[["pts","reb","ast","net_rating","gp"]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(stats, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between player stats")
plt.show()
```
