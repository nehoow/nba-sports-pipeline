```python
import pandas as pd

print("Libraries loaded!")
```

```python
df=pd.read_csv("all_seasons.csv")
print ("Data loaded successfully")
print (f"Shape: {df.shape}")
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset=["player_name","pts","reb","ast"])
print(f"Clean data sweep: {df.shape}")
```

```python
df = df.drop_duplicates()
print(f"After dropping duplicates: {df.shape}")
```

```python
df_recent = df[df["season"] >= "2015-16"]
print(f"Recent Seasons: {df_recent.shape}")
```

```python
player = df[df["player_name"] == "LeBron James"]
print(player[["player_name","season","pts","reb","ast"]])
```

```python
season_avg = df.groupby("season")[["pts", "reb", "ast"]].mean().round(2)
print(season_avg)
```

```python
top_pts = df.groupby("player_name")["pts"].mean().sort_values(ascending=False).head(10)
print(top_pts)
```

```python
top_ast = df.groupby("player_name")["ast"].mean().sort_values(ascending=False).head(10)
print(top_ast)
```

```python
elite = df.groupby("player_name")["pts"].mean()
elite_scorers = elite[elite >= 25].sort_values(ascending=False)
print(elite_scorers)
```

```python
df.to_csv("nba_clean.csv", index=False)
print("Clean data saved as nba_clean.csv!")
```
